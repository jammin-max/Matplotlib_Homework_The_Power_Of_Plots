## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "Resources/Mouse_metadata.csv"
study_results_path = "Resources/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
data_combined_df =pd.merge(mouse_metadata,study_results,how = "outer", on ="Mouse ID")
# Display the data table for preview
data_combined_df.head()
```

```python
# Checking the number of mice.
number_of_mice =data_combined_df["Mouse ID"].count()
number_of_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice=data_combined_df[data_combined_df.duplicated(["Mouse ID","Timepoint"])]

duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

duplicate_mouse_id =data_combined_df[data_combined_df.duplicated(["Mouse ID"])]
duplicate_mouse_id
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data_df=data_combined_df.drop_duplicates("Mouse ID")
```

```python
# Checking the number of mice in the clean DataFrame.
clean_data_df
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
mean = data_combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
median = data_combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
variance = data_combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
standard_dev = data_combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sem = data_combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
# mean, median, variance, standard deviation, and SEM of the tumor volume. 

# Assemble the resulting series into a single summary dataframe.

summary_data_df =pd.DataFrame({"Mean":mean,"Median":median,"Variance":variance,
                               "Standard Deviation":standard_dev,"SEM":sem})
summary_data_df
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of measurements taken on each drug regimen using pandas.
bar_data =pd.DataFrame(data_combined_df.groupby(['Drug Regimen']).count()).reset_index()
#bar_data

bar_1_df =bar_data[["Drug Regimen","Mouse ID"]]
bar_1_df = bar_1_df.set_index("Drug Regimen")
# creating the bar chart of the data
bar_1_df.plot(kind="bar", figsize=(10,5), color = "b", legend= "Trials")

# Set a title for the chart
plt.title("Trials per Drug Regimen")
#Show the bar chart, and utilize tight layout
plt.show()
plt.tight_layout()
```
