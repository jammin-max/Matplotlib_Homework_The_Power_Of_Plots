## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "Resources/Mouse_metadata.csv"
study_results_path = "Resources/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
data_combined_df =pd.merge(mouse_metadata,study_results,how = "outer", on ="Mouse ID")
# Display the data table for preview
data_combined_df.head()
```

```python
# Checking the number of mice.
number_of_mice =data_combined_df["Mouse ID"].count()
number_of_mice
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
mean = data_combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
median = data_combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
variance = data_combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
standard_dev = data_combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sem = data_combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
# mean, median, variance, standard deviation, and SEM of the tumor volume. 

# Assemble the resulting series into a single summary dataframe.

summary_data_df =pd.DataFrame({"Mean":mean,"Median":median,"Variance":variance,
                               "Standard Deviation":standard_dev,"SEM":sem})
summary_data_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
summary_stats = data_combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean().median()
summary_stats_df =pd.DataFrame({"Mean":mean,"Median":median,"Variance":variance,
                               "Standard Deviation":standard_dev,"SEM":sem})
summary_stats_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
data_combined_df
# Start by getting the last (greatest) timepoint for each mouse
# Display the data table for preview
# Calculate the IQR and quantitatively determine if there are any potential outliers. 

# Locate the rows which contain mice on each drug and get the tumor volumes
Capomulin_df = data_combined_df.loc[data_combined_df["Drug Regimen"] == "Capomulin",:]
Ramicane_df = data_combined_df.loc[data_combined_df["Drug Regimen"] == "Ramicane", :]
Infubinol_df = data_combined_df.loc[data_combined_df["Drug Regimen"] == "Infubinol", :]
Ceftamin_df = data_combined_df.loc[data_combined_df["Drug Regimen"] == "Ceftamin", :]

# Capomulin
Capomulin_last = Capomulin_df.groupby('Mouse ID').max()['Timepoint']
Capomulin_vol = pd.DataFrame(Capomulin_last)
Capomulin_merge = pd.merge(Capomulin_vol, data_combined_df, on=("Mouse ID","Timepoint"),how="left")
Capomulin_merge.head()
```

```python
# Capomulin quartiles and IQR
# Determine outliers using upper and lower bounds
Capomulin_tumors = Capomulin_merge["Tumor Volume (mm3)"]

quartiles =Capomulin_tumors.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq


print(f"The lower quartile of Capomulin tumors: {lowerq}")
print(f"The upper quartile of Capomulin tumors: {upperq}")
print(f"The interquartile range of Capomulin tumors: {iqr}")
print(f"The median of Capomulin tumors: {quartiles[0.5]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)

print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
corr=round(st.pearsonr(avg_capm_vol['Weight (g)'],avg_capm_vol['Tumor Volume (mm3)'])[0],2)
print(f"The correlation between mouse weight and average tumor volume is {corr}")
```

```python

# Add the linear regression equation and line to plot
Capomulin_df = data_combined_df.loc[(data_combined_df["Drug Regimen"] == "Capomulin"),:]
# Pull values for x and y values
mouse_weight = Capomulin_df.groupby(Capomulin_df["Mouse ID"])["Weight (g)"].mean()
tumor_volume = Capomulin_df.groupby(Capomulin_df["Mouse ID"])["Tumor Volume (mm3)"].mean()

# Perform a linear regression on year versus violent crime rate
slope, int, r, p, std_err = st.linregress(mouse_weight, tumor_volume)
      
# Create equation of line to calculate predicted violent crime rate
fit = slope * mouse_weight + int

# Plot the linear model on top of scatter plot 
plt.scatter(mouse_weight,tumor_volume)
plt.xlabel("Weight of Mouse")
plt.ylabel("Tumor Volume")
plt.plot(mouse_weight,fit,"--")
plt.xticks(mouse_weight, rotation=90)
plt.show()

```
